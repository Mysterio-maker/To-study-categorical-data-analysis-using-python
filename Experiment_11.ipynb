```python
# Nominal Data ---> categories without any order
# Example:- Gender, ethnicity
# Ordinal Data ---> categories with order
# Example:- grades, income
```

```python
# Using first dataset
import pandas as pd
data = {
'Order_ID':['01','02','03','04','05','06','07','08','09','10'],
'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],
'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],
'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],
'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}
df = pd.DataFrame(data)
print(df)

# Frequency Count----> counts how many orders are there in each category.
print(df['Category'].value_counts())
print()
print(df['Payment_Method'].value_counts())

# Unique Values--->displays all unique categories
print(df['Category'].unique())
print()
print(df['Delivery_Type'].unique())

# Number of unique values--->how many unique values are there
print(df['Category'].nunique())

# Cross tabulation---->category vs. payment method--->relationship between two variables
print(pd.crosstab(df['Category'], df['Payment_Method']))
print()
print(pd.crosstab(df['Delivery_Type'], df['Customer_Type']))

# Percentage distribution ---> tells about the weightage of every quantity in terms of its percentage
print(df['Category'].value_counts(normalize=True)*100)

# Filtering Data ---> to access specific data containing common variable
# Example: orders from electronics category

electronics_orders = df[df['Category']=='Electronics']
print(electronics_orders)

# Grouping ----> combining ---> orders by category and payment method
print(df.groupby('Category')['Payment_Method'].value_counts())

# Sorting Data ---> arranges data in ascending and alphabetical order
sorted_df = df.sort_values(by='Category')
print(sorted_df)
```

```python
# Using second dataset

import pandas as pd

df = pd.read_csv("/content/Expt11.csv")
print(df.head())

# Count students in each grade
print("Grade Freuency:")
print(df['Grade'].value_counts())

# Percentage Distribution of Grades
print("Grade Percentage:")
print(df['Grade'].value_counts(normalize=True)*100)

# Number of students in each department
print("Number of Students, Department-wise:")
print(df['Department'].value_counts())

# Gender distribution
print("Gender count:")
print(df['Gender'].value_counts())

# Gender vs. Grade Analysis
print("Gender vs. Grade:")
print(pd.crosstab(df['Gender'], df['Grade']))

print()

print("Department vs. Gender:")
print(pd.crosstab(df['Department'], df['Gender']))

# Grouping Data
print(df.groupby('Department')['Grade'].value_counts())
```
